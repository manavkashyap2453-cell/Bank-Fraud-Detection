# **Import required libraries**

```python
import joblib

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score
)
```

## Load the training artifacts

```python
import os
import joblib


def load_training_artifacts(artifact_dir):
    """
    Load all artifacts required for model training.
    """

    artifacts = {
        "preprocessor": "preprocessor.pkl",
        "X_train_processed": "X_train_processed.pkl",
        "X_test_processed": "X_test_processed.pkl",
        "X_train_balanced": "X_train_balanced.pkl",
        "y_train": "y_train.pkl",
        "y_test": "y_test.pkl",
        "y_train_balanced": "y_train_balanced.pkl",
    }

    loaded_artifacts = {}

    for key, filename in artifacts.items():
        filepath = os.path.join(artifact_dir, filename)
        loaded_artifacts[key] = joblib.load(filepath)

    return loaded_artifacts
```

```python
artifacts = load_training_artifacts("artifacts")

preprocessor = artifacts["preprocessor"]

X_train_processed = artifacts["X_train_processed"]
X_test_processed = artifacts["X_test_processed"]

X_train_balanced = artifacts["X_train_balanced"]

y_train = artifacts["y_train"]
y_test = artifacts["y_test"]
y_train_balanced = artifacts["y_train_balanced"]
```

```python
print(X_train_processed.shape)
print(X_test_processed.shape)

print(X_train_balanced.shape)

print(y_train.shape)
print(y_test.shape)
print(y_train_balanced.shape)
```

# **Logistic regression**

```python
log_reg = LogisticRegression(
    random_state=42,
    max_iter=1000
)

log_reg.fit(X_train_processed, y_train)
```

```python
y_pred = log_reg.predict(X_test_processed)
y_pred_proba = log_reg.predict_proba(X_test_processed)[:, 1]
```

```python
print(f"Accuracy : {accuracy_score(y_test, y_pred):.4f}")
print(f"Precision: {precision_score(y_test, y_pred):.4f}")
print(f"Recall   : {recall_score(y_test, y_pred):.4f}")
print(f"F1 Score : {f1_score(y_test, y_pred):.4f}")
print(f"ROC AUC  : {roc_auc_score(y_test, y_pred_proba):.4f}")
```

**It confirms two things:**

*  Class imbalance is a major issue.
*  A plain Logistic Regression on the original data is not suitable for this problem.

```python
log_reg_smote = LogisticRegression(
    random_state=42,
    max_iter=1000
)

log_reg_smote.fit(X_train_balanced, y_train_balanced)
```

```python
y_pred_smote = log_reg_smote.predict(X_test_processed)
y_pred_proba_smote = log_reg_smote.predict_proba(X_test_processed)[:, 1]
```

```python
print(f"Accuracy : {accuracy_score(y_test, y_pred_smote):.4f}")
print(f"Precision: {precision_score(y_test, y_pred_smote):.4f}")
print(f"Recall   : {recall_score(y_test, y_pred_smote):.4f}")
print(f"F1 Score : {f1_score(y_test, y_pred_smote):.4f}")
print(f"ROC AUC  : {roc_auc_score(y_test, y_pred_proba_smote):.4f}")
```

CONCLUSION

*  SMOTE helped the model pay attention to the minority class, but it did not make Logistic Regression a good classifier for this dataset.

*  This matches your EDA observation:
There are no strong linear relationships among the numerical features.

Logistic Regression learns a linear decision boundary. If the fraud patterns are complex and non-linear, simply balancing the data won't make a linear model perform well.

# **Decision Tree**

```python
decision_tree = DecisionTreeClassifier(
    random_state=42
)

decision_tree.fit(X_train_processed, y_train)
```

```python
y_pred_dt = decision_tree.predict(X_test_processed)
y_pred_proba_dt = decision_tree.predict_proba(X_test_processed)[:, 1]
```

```python
print(f"Accuracy : {accuracy_score(y_test, y_pred_dt):.4f}")
print(f"Precision: {precision_score(y_test, y_pred_dt):.4f}")
print(f"Recall   : {recall_score(y_test, y_pred_dt):.4f}")
print(f"F1 Score : {f1_score(y_test, y_pred_dt):.4f}")
print(f"ROC AUC  : {roc_auc_score(y_test, y_pred_proba_dt):.4f}")
```

```python
decision_tree_smote = DecisionTreeClassifier(
    random_state=42
)

decision_tree_smote.fit(
    X_train_balanced,
    y_train_balanced
)
```

```python
y_pred_dt_smote = decision_tree_smote.predict(X_test_processed)
y_pred_proba_dt_smote = decision_tree_smote.predict_proba(X_test_processed)[:, 1]
```

```python
print(f"Accuracy : {accuracy_score(y_test, y_pred_dt_smote):.4f}")
print(f"Precision: {precision_score(y_test, y_pred_dt_smote):.4f}")
print(f"Recall   : {recall_score(y_test, y_pred_dt_smote):.4f}")
print(f"F1 Score : {f1_score(y_test, y_pred_dt_smote):.4f}")
print(f"ROC AUC  : {roc_auc_score(y_test, y_pred_proba_dt_smote):.4f}")
```

Unlike Logistic Regression, SMOTE did not help the Decision Tree. In fact, every metric became slightly worse.

This isn't unusual. An unrestricted Decision Tree is a high-variance model and can easily overfit, especially to synthetic samples generated by SMOTE.

At this point, I can conclude two things:

*  Logistic Regression struggles because it cannot model the underlying patterns.
*  A single Decision Tree is too unstable and does not generalize well.

# **Random Forest**

```python
random_forest = RandomForestClassifier(
    random_state=42,
    n_estimators=100
)

random_forest.fit(
    X_train_processed,
    y_train
)
```

```python
y_pred_rf = random_forest.predict(X_test_processed)
y_pred_proba_rf = random_forest.predict_proba(X_test_processed)[:, 1]
```

```python
print(f"Accuracy : {accuracy_score(y_test, y_pred_rf):.4f}")
print(f"Precision: {precision_score(y_test, y_pred_rf):.4f}")
print(f"Recall   : {recall_score(y_test, y_pred_rf):.4f}")
print(f"F1 Score : {f1_score(y_test, y_pred_rf):.4f}")
print(f"ROC AUC  : {roc_auc_score(y_test, y_pred_proba_rf):.4f}")
```

```python
random_forest_balanced = RandomForestClassifier(
    random_state=42,
    n_estimators=100,
    class_weight="balanced"
)

random_forest_balanced.fit(
    X_train_processed,
    y_train
)
```

```python
y_pred_rf_balanced = random_forest_balanced.predict(X_test_processed)
y_pred_proba_rf_balanced = random_forest_balanced.predict_proba(X_test_processed)[:, 1]


print(f"Accuracy : {accuracy_score(y_test, y_pred_rf_balanced):.4f}")
print(f"Precision: {precision_score(y_test, y_pred_rf_balanced):.4f}")
print(f"Recall   : {recall_score(y_test, y_pred_rf_balanced):.4f}")
print(f"F1 Score : {f1_score(y_test, y_pred_rf_balanced):.4f}")
print(f"ROC AUC  : {roc_auc_score(y_test, y_pred_proba_rf_balanced):.4f}")
```

#**Conclusion**

Multiple classification models were trained and evaluated using different strategies for handling class imbalance, including baseline training, SMOTE, and class weighting.

The experimental results showed that:

* Logistic Regression trained on the original dataset failed to identify fraudulent transactions due to the severe class imbalance.
* Applying SMOTE improved the recall of Logistic Regression but resulted in low precision and did not improve the model's overall discriminative ability.
* Decision Tree models exhibited only marginal predictive performance, with SMOTE providing no measurable benefit.
* Random Forest models, both with and without class weighting, also failed to distinguish fraudulent transactions effectively.

Across all experiments, the ROC-AUC remained close to **0.50**, indicating that the trained models were unable to reliably separate fraudulent and legitimate transactions.

Based on these observations, no model demonstrated satisfactory predictive performance on the available dataset. The results suggest that the features provided in the dataset contain limited predictive information for the target variable, making fraud detection challenging despite the use of feature engineering, preprocessing, and imbalance handling techniques.
